```
import os
import nibabel as nib                                                                                                              

def list_files(dir): 
    i=0
    j=0                                                                                                 
    r = []                                                                                                            
    subdirs = [x[0] for x in os.walk(dir)]                                                                            
    for subdir in subdirs:
        i=i+1 
        j=j+1                                                                                           
        files = os.walk(subdir).__next__()[2]                                                                             
        if (len(files) > 0):                                                                                          
            for file in files:
              img = nib.load(subdir+'/'+file)
              if file == 'segmentation.nii.gz':
                nib.save(img, os.path.join('segmentation','segmentation_' + str(i) + '.nii.gz'))
          
              if file == 'imaging.nii.gz':
                nib.save(img, os.path.join('imaging','imaging_' + str(j) + '.nii.gz'))
            
            
                                                                                               
            
```
