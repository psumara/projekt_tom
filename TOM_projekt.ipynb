```
import os
import nibabel as nib                                                                                                              

def list_files(dir): 
    i=-2
    j=-2                                                                                                
                                                                                                    
    subdirs = [x[0] for x in os.walk(dir)]                                                                            
    for subdir in subdirs:
        i=i+1 
        j=j+1                                                                                           
        files = os.walk(subdir).__next__()[2]                                                                             
        if (len(files) > 0):                                                                                          
            for file in files:              
              if file == 'segmentation.nii.gz':
                os.rename(subdir+'/'+file, 'segmentation/segmentation_' + str(i).zfill(3) + '.nii.gz')
              if file == 'imaging.nii.gz':
                os.rename(subdir+'/'+file, 'imaging/imaging_' + str(j).zfill(3) + '.nii.gz')
```
