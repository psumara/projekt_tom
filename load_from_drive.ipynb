```
import tensorflow as tf
tf.compat.v1.disable_eager_execution()
```

```
from image_preprocessing import *
```

```
X_train = imaging_preprocessing('train/imaging')
y_train = segmentation_preprocessing('train/segmentation')
X_test = imaging_preprocessing('test/imaging')
y_test = segmentation_preprocessing('test/segmentation')
```

```
train_set_size = len(X_train.filenames)
test_set_size = len(X_test.filenames)
train_set = zip(X_train, y_train)
test_set = zip(X_test, y_test)
```

```
from unet import *
```

```
IMG_WIDTH=256
IMG_HEIGHT=256
IMG_CHANNELS=1
inputs = tf.keras.layers.Input((IMG_WIDTH, IMG_HEIGHT, IMG_CHANNELS))
outputs = unet_core(inputs)
model = tf.keras.Model(inputs = [inputs], outputs = [outputs])
model.compile(optimizer = 'adam', loss = dice_coef_loss, metrics = [dice_coef])
model.summary()
```

```
import numpy as np
callbacks=[tf.keras.callbacks.EarlyStopping(patience=8, monitor='val_loss'), 
            tf.keras.callbacks.TensorBoard(log_dir='/tmp/logs')]

BATCH_SIZE = 32
EPOCHS = 5
results = model.fit(
    train_set,  
    steps_per_epoch= train_set_size // BATCH_SIZE,
    epochs = EPOCHS,
    validation_data = test_set, 
    validation_steps = test_set_size // BATCH_SIZE,
    callbacks = callbacks,
    verbose = 1)
```
